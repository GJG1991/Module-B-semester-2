```python
import matplotlib.pyplot as plt

labels = ['Supplier A', 'Supplier B', 'Supplier C', 'Supplier D']
sizes = [34, 31, 9, 26]
colors = ['#003366', '#336699', '#99CCFF', '#6699CC'] 
explode = (0.05, 0.05, 0.05, 0.05)

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(sizes, 
                                  explode=explode, 
                                  labels=None, 
                                  autopct='%1.0f%%',
                                  shadow=True, 
                                  startangle=140, 
                                  colors=colors, 
                                  textprops={'fontsize': 14, 'color': 'white'})

autotexts[2].set_color('black') 
autotexts[3].set_color('black')

ax.legend(wedges, labels,
          title="",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          frameon=False,
          fontsize=12)

plt.title("Supplier Market Share", fontsize=16, loc='left', pad=20, color='#444444')
plt.axis('equal')

plt.tight_layout()
plt.show()
```

# faf5 dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_faf5 = pd.read_csv('datasets/faf5.csv')
df_faf5.head()
```

```python
df_faf5.info()
```

```python
fig, axes = plt.subplots(1,2, figsize=(15, 6))

df_faf5['tons_2023'].plot.hist(bins=30, ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Tons (Normal Scale)')

df_faf5['tons_2023'].plot.hist(bins=30, ax=axes[1], color='salmon', edgecolor='black')
axes[1].set_yscale('log')
axes[1].set_title('Distribution of Tons (Log Scale)')

plt.show()
```

```python
df_faf5['dms_mode'].value_counts()
```

# changed the dms_mode column to a more readable format using a dictionary mapping, and created a new column called mode_name to store the mapped values. This allows for easier interpretation of the transportation modes in the dataset.

```python
mode_lookup = {
    1: 'Truck',
    2: 'Rail',
    3: 'Water',
    4: 'Air',
    5: 'Multiple/Mail',
    6: 'Pipeline',
    7: 'Other',
    8: 'Import/Export'
}

df_faf5['mode_name'] = df_faf5['dms_mode'].map(mode_lookup)

df_faf5['mode_name'].value_counts()
```

```python
tons_by_mode = df_faf5.groupby('mode_name')['tons_2023'].sum()
print(tons_by_mode.apply(lambda x: '{:,.2f}'.format(x)))
```

```python
df_faf5.groupby('mode_name')['current_value_2023'].sum().apply(lambda x: '{:,.2f}'.format(x))
```

# conclusion faf5 dataset
the data was clean after we mapped the mdoes to a more readable format, and we were able to draw conclusions about the distribution of tons and current value across different transportation modes. The distribution of tons was skewed, with a long tail on the right side, indicating that there are some very large shipments. The bar plot showed that the majority of shipments were made by truck, followed by rail and then water. The total value of shipments was also highest for truck, followed by rail and water. Overall, the data provided insights into the transportation patterns and values in the dataset.

#  LaDe dataset

```python
df_delivery = pd.read_csv('datasets/delivery_five_cities.csv')
df_delivery.head()
```

```python
df_delivery.info()
```

```python
df_delivery.columns
```

```python
df_delivery['receipt_time'] = '2018-' + df_delivery['receipt_time']
df_delivery['sign_time'] = '2018-' + df_delivery['sign_time']

df_delivery['receipt_time'] = pd.to_datetime(df_delivery['receipt_time'])
df_delivery['sign_time'] = pd.to_datetime(df_delivery['sign_time'])

print(df_delivery[['receipt_time', 'sign_time']].head())
print(df_delivery.dtypes)
```

```python
df_delivery['duration'] = (df_delivery['sign_time'] - df_delivery['receipt_time']).dt.total_seconds() / 3600
print(df_delivery[['receipt_time', 'sign_time', 'duration']].head())
```

```python
df_delivery['duration'].describe()
```

```python
df_clean_delivery = df_delivery[(df_delivery['duration'] > 0) & (df_delivery['duration'] < 24)]
```

```python
df_clean_delivery['duration'].describe()
```

```python
df_clean_delivery['duration'].plot.hist(bins=30, color='green', edgecolor='black')
plt.title('Distribution of Delivery Times (cleaned)')
plt.xlabel('Hours')
plt.show()
```

# CONCLUSION LaDe dataset
The LaDe dataset appears to be clean and usable for analysis. The distribution of delivery times is right-skewed, with a long tail on the right side, indicating that there are some deliveries that take significantly longer than the majority. The histogram of delivery times shows that most deliveries are completed within a certain range, but there are outliers that take much longer. The boxplot confirms the presence of these outliers and provides a visual representation of the spread of delivery times. Overall, the dataset provides insights into the distribution of delivery times and can be used for further analysis to identify factors that may contribute to longer delivery times.
